# 1. 인구 피라미드
---

## 데이터 가져오기
---
- 전국 2011년 남여 연령별인구현황

```python
import pandas as pd
year = 2022
file_name = "202203_202203_연령별인구현황_월간"

df_m = pd.read_excel(f"{file_name}.xlsx")
df_m
```

## 남자 데이터 가져오기
---

```python
# 남자 데이터
# 위 3 줄 제외
df_m = pd.read_excel(f"{file_name}.xlsx", skiprows=3, index_col="행정기관", usecols="B, E:Y")
df_m.head(3) # 위에서 부터 3개
```

### 남자 전국 데이터 정수화
---
- str인 데이터를 int형으로 바꾼다
- ex) 1,118,000 -> 1118000

```python
# , 대신 빈칸을 넣고 int로 바꾼다.
df_m.iloc[0] = df_m.iloc[0].str.replace(",", "").astype(int)
```

```python
df_m.iloc[0]
```

```python
df_m
```

## 여자 데이터 가져오기
---

```python
# 여자 데이터
# 위 3 줄 제외
df_w = pd.read_excel(f"{file_name}.xlsx", skiprows=3, index_col="행정기관", usecols="B, AB:AV")
df_w.head(3) # 위에서 부터 3개
```

- 앞에 데이터에 0 ~ 4세 데이터가 있어서 0~4.1 이라는 값이 나옴

```python
df_m.columns
```

```python
df_w.columns
```

- column명 통일

```python
df_w.columns = df_m.columns
df_w.columns
```

### 여자 전국 데이터 정수화
---

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(",", "").astype(int)
df_w.iloc[0]
```

```python
df_w
```

## 데이터 시각화
---
- matplotlib 모듈 사용

### 가로 막대 그래프 생성
---

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams["font.family"] = "Malgun Gothic" # windows
matplotlib.rcParams["font.size"] = 15
matplotlib.rcParams["axes.unicode_minus"] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상 해결
```

```python
plt.figure(figsize=(10, 7))
plt.barh(df_m.columns, df_m.iloc[0] // 1000) # 단위 : 천명
plt.barh(df_w.columns, df_w.iloc[0] // 1000) # 단위 : 천명
```

- 남여 그래프가 겹쳐서 나오므로 여자 데이터에 -를 준다.

```python
plt.figure(figsize=(10, 7))
plt.barh(df_m.columns, df_m.iloc[0] // 1000) # 단위 : 천명
plt.barh(df_w.columns, -df_w.iloc[0] // 1000) # 단위 : 천명
```

### 제목 추가
---

```python
plt.figure(figsize=(10, 7))
plt.barh(df_m.columns, df_m.iloc[0] // 1000) # 단위 : 천명
plt.barh(df_w.columns, -df_w.iloc[0] // 1000) # 단위 : 천명
plt.title(f"{year} 대한민국 인구 피라미드")
plt.show()
```

### label 추가
---

```python
plt.figure(figsize=(10, 7))
plt.barh(df_m.columns, df_m.iloc[0] // 1000, label="남자") # 단위 : 천명
plt.barh(df_w.columns, -df_w.iloc[0] // 1000, label="여자") # 단위 : 천명
plt.title(f"{year} 대한민국 인구 피라미드")
plt.legend()
plt.show()
```

### 파일 저장
---

```python
plt.figure(figsize=(10, 7))
plt.barh(df_m.columns, df_m.iloc[0] // 1000, label="남자") # 단위 : 천명
plt.barh(df_w.columns, -df_w.iloc[0] // 1000, label="여자") # 단위 : 천명
plt.title(f"{year} 대한민국 인구 피라미드")
plt.legend()
plt.savefig(f"{year}_인구피라미드.png", dpi=100)
plt.show()
```
