# 9. 데이터 선택(조건)
---
- 조건에 해당하는 데이터 선택

```python
import pandas as pd

df = pd.read_excel("score.xlsx", index_col="지원번호")
df
```

```python
df["키"] >= 185 # 키가 185이상 인지 여부를 True/ False
```

```python
filt = (df["키"] >= 185)
df[filt]
```

```python
df[~filt] # filt를 역으로 적용
```

```python
df[df["키"] >= 185]
```

```python
# df.loc[row , column]
df.loc[df["키"] >= 185, "수학"] # 키가 185이상인 학생의 수학 점수
```

```python
df.loc[df["키"] >= 185, ["이름", "수학", "과학"]] # 키가 185이상인 학생의 이름, 수학, 과학 데이터
```

## 다양한 조건
---
### & 그리고(and)

```python
# 괄호() 필수
df.loc[(df["키"] >= 185) & (df["학교"] == "북산고")] # 키가 185이상 북산고인 학생 데이터
```

### | 또는(or)
---

```python
df.loc[(df["키"] < 170) | (df["키"] > 200)] # 키가 170 미만 또는 키가 200초과인 학생 데이터
```

### str함수
---

```python
# 송으로 시작하는 이름인 학생 데이터
filt = df["이름"].str.startswith("송")
df[filt]
```

```python
# 이름에 태가 들어가는 학생 데이터
filt = df["이름"].str.contains("태")
df[filt]
```

```python
# 이름에 태가 들어가지 않는 학생 데이터
df[~filt]
```

```python
# 리스트를 활용한 데이터 선택
langs = ["Python", "Java"]
# 특기가 파이썬이나 자바인 학생
# 리스트와 값이 일치하는 값만 나온다
filt = df["SW특기"].isin(langs)
df[filt]
```

> PYTHON 인 데이터 값이 제외됨  
> 리스트에 값과 정확히 일치하지 않기떄문

```python
langs = ["python", "java"]
filt = df["SW특기"].str.lower().isin(langs)
df[filt]
```

 - df["SW특기"].str.lower().isin(langs)는 True / False 값으로 리턴함

```python
df["SW특기"].str.contains("Java", na=True) # NaN 데이터에 대해서 True로 설정
```

- NaN 값이 있으면 오류남

```python
df["SW특기"].str.contains("Java", na=False) # NaN 데이터에 대해서 True로 설정
```

```python
filt = df["SW특기"].str.contains("Java", na=False) # NaN 데이터에 대해서 True로 설정
df[filt]
```
