## COVID-19 백신 종류별 접종 인구
---

```python
days = [1, 2, 3] # 일
# 단위 만 명
az = [2, 4, 8] # az 접종 인구
pfizer = [5, 1, 3] # 화이지 접종 인구
moderna = [1, 2 ,5] # 모더나 접종 인구

plt.plot(days, az)
plt.plot(days, pfizer)
plt.plot(days, moderna)
```

```python
plt.plot(days, az, label="az")
plt.plot(days, pfizer, label="화이자", marker="o", ls="--")
plt.plot(days, moderna, label="모더나", marker="s", ls="-.")

plt.legend()
```

```python
plt.plot(days, az, label="az")
plt.plot(days, pfizer, label="화이자", marker="o", ls="--")
plt.plot(days, moderna, label="모더나", marker="s", ls="-.")

plt.legend(ncol=3) # 가로로 3개 표시됨
```
