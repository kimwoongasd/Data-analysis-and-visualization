# 4. 스타일
---

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic" # windows
# matplotlib.rcParams['font family'] = "AppleGothic" # Mac
matplotlib.rcParams["font.size"] = 15
matplotlib.rcParams["axes.unicode_minus"] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상 해결

x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
```

- 선 두껍게

```python
plt.plot(x, y, linewidth=5)
```

## 마커 (marker)
---

```python
plt.plot(x, y, marker="o")
```

```python
plt.plot(x, y, marker="o", linestyle="None")
```

```python
plt.plot(x, y, marker="v")
```

- 마커 사이즈 변경

```python
plt.plot(x, y, marker="v", markersize=10)
```

- 마커 모양 변경

```python
plt.plot(x, y, marker="X", markersize=10)
```

- 마커 테두리 색깔 변경

```python
plt.plot(x, y, marker="X", markersize=10, markeredgecolor="red")
```

```python
plt.plot(x, y, marker="o", markersize=10, markeredgecolor="red")
```

- 마커 색 변경

```python
plt.plot(x, y, marker="o", markersize=10, markeredgecolor="red", markerfacecolor="yellow")
```

## 선 스타일
---

```python
plt.plot(x, y, linestyle=":")
```

```python
plt.plot(x, y, linestyle="--")
```

```python
plt.plot(x, y, linestyle="-.")
```

## 색깔
---

```python
plt.plot(x, y, color="pink")
```

```python
plt.plot(x, y, color="#ff0000")
```

```python
plt.plot(x, y, color="b")
```

```python
plt.plot(x, y, color="g")
```

## 포맷
---
- 간단하게 스타일을 모두 지정 할 수 있다.

```python
plt.plot(x, y, "ro--") # color, marker, linestyle
```

```python
plt.plot(x, y, "bv:")
```

```python
plt.plot(x, y, "go") # linestyle 넣지 않으면 None
```

## 축약어
---

```python
# mfc = markerfacecolor, ms = markersize, mec = markeredgecolor, ls = linestyle
plt.plot(x, y, marker="o", mfc="red", ms=10, mec="blue", ls=":")
```

## 투명도
---

```python
plt.plot(x, y, marker="o", mfc="red", ms=10, alpha=0.3) # alpha : 투명도 (0 ~ 1)
```

## 그래프 크기
---

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y)
```

```python
plt.figure(figsize=(5, 10))
plt.plot(x, y)
```

```python
plt.figure(figsize=(10, 5), dpi=500) # dpi : 해상도
plt.plot(x, y)
```

## 배경색
---

```python
plt.figure(facecolor="yellow")
plt.plot(x, y)
```

```python
plt.figure(facecolor="#0000")
plt.plot(x, y)
```
