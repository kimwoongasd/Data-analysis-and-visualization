```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic" # windows
# matplotlib.rcParams['font family'] = "AppleGothic" # Mac
matplotlib.rcParams["font.size"] = 15
matplotlib.rcParams["axes.unicode_minus"] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상 해결

x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
# fontdict을 통해서 따로 글꼴과 사이즈 변경 가능
plt.title("꺽은선 그래프", fontdict={"family" : "HYGungSo-Bold", "size":20})
```

- x, y 축 label 추가

```python
plt.plot(x, y)
plt.xlabel("x축")
plt.ylabel("y축")
```

- x, y축 label 색깔 설정

```python
plt.plot(x, y)
plt.xlabel("x축", color="red")
plt.ylabel("y축", color="#00aa00")
```

- x, y축 label 위치 설정

```python
plt.plot(x, y)
plt.xlabel("x축", color="red", loc="right") # right, left, center
plt.ylabel("y축", color="#00aa00", loc="top") # top, center, bottom
```

- x, y축 tick 설정

```python
plt.plot(x, y)
plt.xticks([1, 2, 3])
plt.yticks([3, 6, 9, 12])
plt.show()
```
