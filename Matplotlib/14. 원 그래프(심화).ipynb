```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = "Malgun Gothic" # windows
# matplotlib.rcParams['font family'] = "AppleGothic" # Mac
matplotlib.rcParams["font.size"] = 15
matplotlib.rcParams["axes.unicode_minus"] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상 해결

values = [30, 25, 20, 13, 10, 2]
labels = ["Python", "Java", "Javascript", "C#", "C/C++", "ETC"]
```

```python
plt.pie(values, labels=labels, autopct="%.1f%%", startangle=90, counterclock=False)
plt.show()
```

### 색 변경
---

```python
values = [30, 25, 20, 13, 10, 2]
labels = ["Python", "Java", "Javascript", "C#", "C/C++", "ETC"]
# colors = ["b", "g", "r", "c", "m", "y"]
colors = ["#ffadad", "#ffd6a5", "#ddffb6", "#caffdf", "#9bf6ff", "#a0c4ff"]
explode = [0.05] * 6

plt.pie(values, labels=labels, autopct="%.1f%%", startangle=90, counterclock=False, colors=colors, explode=explode)
plt.show()
```

### 속성 정의
---

- 도넛 모양 만들기

```python
wedgeprops = {"width":0.8}
plt.pie(values, labels=labels, autopct="%.1f%%", startangle=90, counterclock=False, colors=colors, wedgeprops=wedgeprops)
plt.show()
```

- 간격 주기

```python
wedgeprops = {"width":0.8, "edgecolor":"w", "linewidth":3}
plt.pie(values, labels=labels, autopct="%.1f%%", startangle=90, counterclock=False, colors=colors, wedgeprops=wedgeprops)
plt.show()
```

```python
# 10퍼센트 이상만 보여주는 함수
def custom_autopct(pct):
    return ("%.1f%%" % pct) if pct >= 10 else ""

plt.pie(values, labels=labels, autopct=custom_autopct, startangle=90, counterclock=False, colors=colors, wedgeprops=wedgeprops)
plt.show()
```

### 중심으로 부터 글 사이 간격 조절
---

```python
plt.pie(values, labels=labels, autopct=custom_autopct, startangle=90, counterclock=False, colors=colors, wedgeprops=wedgeprops, pctdistance=0.5)
plt.show()
```
