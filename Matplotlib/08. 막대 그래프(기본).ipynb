# 8. 막대 그래프(기본)
---

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic" # windows
# matplotlib.rcParams['font family'] = "AppleGothic" # Mac
matplotlib.rcParams["font.size"] = 15
matplotlib.rcParams["axes.unicode_minus"] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상 해결

labels = ["강백호", "서태웅", "정대만"] # 이름
values = [190, 187, 184] # 키

# 막대 그래프
plt.bar(labels, values)
```

- 데이터 색 추가

```python
labels = ["강백호", "서태웅", "정대만"] # 이름
values = [190, 187, 184] # 키

# 막대 그래프
plt.bar(labels, values, color="r")
```

- 각 데이터의 색 넣기

```python
labels = ["강백호", "서태웅", "정대만"] # 이름
values = [190, 187, 184] # 키
colors = ["r", "g", "b"]

# 막대 그래프
plt.bar(labels, values, color=colors, alpha=0.5)
```

- y축 값 조정

```python
labels = ["강백호", "서태웅", "정대만"] # 이름
values = [190, 187, 184] # 키
colors = ["r", "g", "b"]

# 막대 그래프
plt.bar(labels, values, color=colors)
plt.ylim(175, 195)
```

- 막대 크기 조절

```python
# 막대 그래프
plt.bar(labels, values, color=colors, width=0.5)
```

- 데이터 기울기

```python
# 막대 그래프
plt.bar(labels, values, color=colors, width=0.5)
plt.xticks(rotation=45) # x 축의 이름 데이터의 각도를 45도로 설정
plt.yticks(rotation=45) # y 축의 이름 데이터의 각도를 45도로 설정
```

- x 축의 이름 변경

```python
labels = ["강백호", "서태웅", "정대만"] # 이름
values = [190, 187, 184] # 키
ticks = ["1번학생", "2번학생", "3번학생"]

plt.bar(labels, values, color=colors, width=0.5)
plt.xticks(labels, ticks)
```
